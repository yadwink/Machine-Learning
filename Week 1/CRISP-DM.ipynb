## Methodology for organising ML projects :CRISP-DM

* Ml projects follow a certain order:

    * Understand the problem
    * Collect the data
    * Train the model
    * Use it

* Methodologies like CRISP-DM helps us to organise ML projects in such an order.

* CRISM-DM is a cross industry standard process for data mining which is a methodology that helps us to organise ML projects.


##### How the spam email detection example can be mapped using CRISP-DM?

 * Step 1: Business understanding -> identifying the problem and understand how to solve it
 
     * User complaints about the spam
     * Analyse to what extent is the problem
     * DO WE REALLY NEED ML FOR THIS PROBELM?
     * If not: propose an alternative solution
     
       *If we decided to go on with the ML solution, then follow the steps:
             
              Define the goal:
              
              * Is the goal to reduce the amount of spam messages, or
              * Is it to reduce the amount of complaints about spam
              
             The goal has to be measurable, e.g., Reduce the amount of spam by 50%

* Step 2: Data understanding -> making sure first that we have data availability. IF THERE IS NO DATA; NO ML CAN BE APPLIED. Thus, analyze available data sources and then decide if we need to get more data

    Data understanding in case of spam email detection example.
    
        Identify the data sources
        
        * We have a report spam button
        * Is the data behind this button any good?
        * Is it reliable? 
        * Do we track it correctly?
        * Is the dataset large enough?
        * Do we need to get more data?
        
        ** identifying data sources step may influence our main goal.Because if we find out there is not enough data or if our problem is different, this may influence our main objective. Therefore, if we get a new information about the data source, then we may go back to the previous step and revise everything.
    

* Step 3: Data preparation -> after learning in previous steps that when we have sufficient data, it is relaible then it can be further used. Thus, in this step we will transform the data so that it can be used in a ML algorithm. This means we will extract different features from the data.

    further steps in data preparations:
    
    * Clean the data. Remove all noise. e.g., when user accidently marks a good email as spam, then we need to clean it.
    
    * Build pipelines. e.g., build sequence of steps which take in raw data, transform, and then clean it.
    
    * Convert the data into the tabular form so it can be used for the ML algorithm. 
    
    

* Step 4: Modelling -> once we have the data is the desired format, we apply the ML model to it. This is the step where actual ML happens.

    * We try different ML models and select the best one
    * There are many different kinds of ML models e.g.,:
        * Logisitic regression
        * Decision tree
        * Neural network
        * or many others
        
    * sometimes we might need to go back to the previous step of data preparation in order to add new features or fix any data related issues.


* Step 5: Evaluation -> this step measures how well the model solves the business problem.
 We will ask following questions:
 
 Is the model good enough?
 
     * Have we reached the goal? 
     * The goal was to reduce the amount of spam emails. Have we reduced it?
     * if yes, then by how much? Is it 50% or 30%? Is 30% good enough?
     * We need to do evaluation on a test group
     * Do our metrics improve?
     
     
     * Maybe at this stp we realise that the set goal was not achievable. Then we need to do a retrospective:
     
          * Was the goal achievable?
          * Did we solve/measure the right thing?
          
      After evaluating above points, we may decide to:
      
          * Go back and adjust the goal
          * Roll the model to all users
          * Stop working on the project etc.
      


* Step 6: Deployment ->  often evaluation and deployment steps go together. 

    * first, we will evaluate the model on few of the user for example, only 5% of the users. If it works well with them then we roll out the model to rest of the users.
    
    * Second, we actually deploy the model
    
    * Thus, roll the model to all users,
    * propoer monitoring of the model 
    ** ensuring the best quality and maintainability
    
* The last step is ITERATION -> ML projects often require multiple iterations. We might need to go back to step 1 of business understanding. It may be that the model is good enough for the moment however, will need improvement few months later. 

* What we can learn from this step: Keep the steps simple, so we can learn from feedback and can easily improve. In this way, iterations are small and things can be done faster.

In summary, we need follwoing steps to organise a ML project:
  * Business understanding -> define a measurable goal and ask if ML is really needed
  * Data understanding -> do we have enough data? DO we need different type or more volume of data
  * Data preparation -> transforming data into tabular form, compatible for ML model
  * Modelling -> where actual ML takes place. Select the best ML model
  * Evaluation -> validate if the set goal is reached or not, otherwise go back to business understanding and do a retrospect
  * Deployment -> roll out the ML model to all users
  * Iterate-> keep it simple, learn from feedback, improve
