## Machine learning model selection process

In the last lesson, we learnt that a methodology (CRISP-DM) to organise a ML model is a seven steps process. The fourth step of modelling is the step where actual ML takes place. The selection of the best ML model depends on the best accuracy achieved by the model.

For example, if we train a ML model on the spam email detection problem dataset in July, we ge the following parameters:

     * X is the parameter matrix which we feed to the ML model
     * y is the target vector, and
     * g is the trained ml model

now if we apply this ML model 'g' to an email in August then it will produce a score that there is 0.7 probability that this email is spam. This is how ML models are applied in practice.



But how to evaluate that how well our ML model will perform on the dataset being applied in upcoming next months? Ofcourse, we can't go in future and see how well the model will perform. But what is possible is that if we divide the dataset into 80:20 ratio and only use the 80% of the data for our model training:

 + Training set -> The 80% part of the data which is being used for ML model training is referred to as TRAINING SET
 + Validation set -> The 20% part of the data which we keep for the later and do not use during the training process, is referred as VALIDATION SET.

Thus, at the end we use feature matrx X and target vector y for both training and validation datasets. 

The validation dataset is not used in training. There are feature matrices and y vectors for both training and validation datasets. The model is fitted with training data, and it is used to predict the y values of the validation feature matrix. Then, the predicted y values (probabilities) are compared with the actual y values.

The feature matrix X is extracted from the training dataset. the prediction vector y and the training model g remain same.
We will extract another feature matrix from the validation dataset, and call is Xv which was not used in the training set. We will also have yv and now we will apply the model from the training dataset to the validation feature matrix Xv to obtain prediction y^:

g(Xv)=y^



The next step would be to compare both predicitons,i.e., yv & y^

The output of y^ are in terms of probablitiy -> 0.8 0.7 0.6 0.1 0.9 0.6  and 
the output of yv are  0 (not spam) & 1 (spam) -> 1   0   1   0   1   0

y^ are our predicitons and yv is our target variable. When making a direct comparison, we can see that 4/6 cases are correct. In this case the model is 66% correct.

Now if performed this operation with different models as:

 * g1 -> Logistic regression and let's say it has 66% accuracy,
 * g2 -> Decision trees and it has 60% accuracy,
 * g3 -> Random forest has 67%, and
 * g4 -> Neural network has 80% accuracy
 
 in this example, the neural network model has the highest accuracy and thus, we will select the neural network as our best model.


However, this are not that straightforward. If we apply multiple different kinds of ML models to the validation dataset, there is a chance that maybe one of the model will produce good result. But this occurs only by chance, we can't take and trust the results from this case. This problem in particular in statistics in known as Multiple Comparions Problem. This problem occurs frequently in ML projects. The reason is that these methdos are probabilitisc and hence, a model might get lucky and produces favourable results.

To guard against such problems we can hold out more than one dataset before training the model. This means instead of only taking 20% data for later validation, we will divide the data into three parts:

 + 20% for the validation 
 + 20% for the test, and
 + 60% for the training
 
 But this division is not always need to be 20,20,60. This division can vary depending on the problem and dataset.
 
 

Now, we will keep the 20% test dataset away and work with our 60% training and 20% validation set.

We perform model selection as described previously, we take X and y and train our model g with the training set, then we use the trained g model and apply it to the Xv and yv validation set and calculate the accuracy.

After selecting our best performed model, we will apply this model to the test dataset which we kept aside in the begining.

We will extract the feature matrix Xt and target varible yt from the test dataset and we will perfrom an extra round of validation using the model g on the test dataset. This test dataset opertaion will also result in an accuracy score. If this score is close to the validation dataset accuracy score, then we can finalise our ML model selection process.


Thus, to summarise the ML model selection process first,:
    
+ split the dataset into three parts: Train, test, and validation set, then
+ train the model
+ validate the model using validation dataset and note the accuracy score
+ this process of training and validation can be repeated on various models 
+ select the best out of them based on the highest accuracy score
+ apply the best selected model to the test data and check for the accuracy
+ finalise the ML model

At the end we see that we kind of throw away the validation set. Therefore, we will combine the 60% train an 20 % validation set and use them as combined set for training. Then we train a model g and apply it to the test set. Thus, in this case we will have more data for training.
